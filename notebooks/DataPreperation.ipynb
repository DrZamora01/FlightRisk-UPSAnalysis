```python
#load raw data file
airbus_df = pd.read_csv(r"../data/raw/AIDS_REPORTS AirBus.csv")
boeings_df = pd.read_csv(r"../data/raw/AIDS_REPORTS Boeings.csv")
md11_df = pd.read_csv(r"../data/raw/AIDS_REPORTS MD_11.csv")
```

```python
#filter for ups
airbus_ups = airbus_df[airbus_df['Operator'] == "UNITED PARCEL SERVICE CO"]
boeings_ups = boeings_df[boeings_df['Operator'] == "UNITED PARCEL SERVICE CO"]
md11_ups = md11_df[md11_df['Operator'] == "UNITED PARCEL SERVICE CO"]
```

```python
airbus_ups["AircraftType"] = "Airbus"
boeings_ups["AircraftType"] = "Boeing"
md11_ups["AircraftType"] = "MD-11"
```

```python
#combined
combined_df = pd.concat([airbus_ups, boeings_ups, md11_ups], ignore_index=True)
```

```python
# Drop columns that are empty or irrelevant
columns_to_drop = [' ', ' .1']
combined_df.drop(columns=columns_to_drop, axis=1, inplace=True, errors='ignore')
```

```python
combined_df.to_csv("../data/processed/UPS_Incidents_Cleaned.csv", index=False)

```
